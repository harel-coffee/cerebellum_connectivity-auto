## Evaluation of Ridge-regression models of cortical-cerebellar connectivity

```python
import seaborn as sb
import pandas as pd
import numpy as np
from connectivity.constants import Dirs
```

```python
d= Dirs()
```

```python
# Load the evalution on the Ridge WB162
T = pd.read_csv(d.CONN_EVAL_DIR / 'Ridge_WB162.dat')
T.columns
```

```python
sb.lineplot(x='logalpha',y='R',data=T)
```

```python
T['noiseceiling_Y']=np.sqrt(T.noise_Y_R)
T['noiseceiling_XY']=np.sqrt(T.noise_Y_R)*np.sqrt(T.noise_X_R)

ax=sb.lineplot(x='logalpha',y='R',data=T)
ax=sb.lineplot(x='logalpha',y='noiseceiling_XY',data=T)
ax.axhline(np.mean(T.noiseceiling_Y))
```

```python
table = pd.pivot_table(T[T.eval_exp==1], values='noise_Y_R', index=['SN'],columns=['name'], aggfunc=len)
table
```
